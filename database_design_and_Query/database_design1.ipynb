```python
# Step three: Invoke callproc to call the stored procedure.

# Call the stored procedure with its name
cursor.callproc("PeakHour")

# Retrieve records in "dataset"
results = next(cursor.stored_results())
dataset = results.fetchall()

# Retrieve column names using list comprehension in a 'for' loop 
for column_name in cursor.stored_results():
    columns = [column[0] for column in column_name.description]

# Print column names
print(columns)

# Print data in a loop
for data in dataset:
    print(data)
```

```python
# Task 3: Implement a stored procedure GuestStatus

# Step one: Write a SQL CREATE PROCEDURE query for GuestStatus. Note that here, there are multiple statements in one Procedure.

# Implementing stored procedure query
print("Creating a stored procedure called 'GuestStatus'")
stored_procedure_query2="""
CREATE PROCEDURE GuestStatus()

BEGIN

SELECT CONCAT(Bookings.GuestFirstName, ' ', Bookings.GuestLastName) as GuestName
FROM Bookings;

SELECT EmployeeID, Role,

CASE
WHEN EmployeeID IN (1, 2) THEN "Ready to pay" 
WHEN EmployeeID IN (3) THEN "Ready to serve" 
WHEN EmployeeID IN (4) THEN "Preparing Order" 
WHEN EmployeeID IN (5) THEN "Order served" 
ELSE "In Queue" 
END AS Status
FROM Employees;

SELECT * FROM Bookings
LEFT JOIN
Employees ON Bookings.EmployeeID=Employees.EmployeeID;

END

"""

# Execute the query : Step five: Run the stored procedure query by invoking execute module on the cursor.
cursor.execute(stored_procedure_query2)
print("Stored procedure 'GuestStatus' creation executed successfully.")
```

```python
# Step six: Invoke callproc to call the stored procedure.

# Call the stored procedure with its name
cursor.callproc("GuestStatus")

# Retrieve recrods in "dataset" - Step seven: Fetch the results in a variable called dataset.
results = next( cursor.stored_results())
dataset = results.fetchall()


# Step eight: Extract the names of the columns.
# Retrieve column names using list comprehension in a 'for' loop 
for column_id in cursor.stored_results():
    columns = [ column[0] for column in column_id.description ]


#Step nine: Print the names of the columns.
# Print column names
print(columns)


#Step ten: Print the sorted data using for loop.
# Print data 
for data in dataset:
    print(data)
```
